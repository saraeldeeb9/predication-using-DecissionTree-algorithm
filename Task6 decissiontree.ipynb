**Lets first import the libraries we will work on:**

```python
# Imports
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Load the data and create dataframe

```python
# Load Data
iris = load_iris()
```

```python
# Create a dataframe
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
```

```python
df.head()
```

```python
# Let's see target names
targets = iris.target_names
print(targets)
```

### Train the data to build the classification model

```python
# Prepare training data for building the model
X_train = df.drop(['target'], axis=1)
y_train = df['target']
```

```python
# Instantiate the model
cls = DecisionTreeClassifier()

# Train/Fit the model 
cls.fit(X_train, y_train)
```

### Now let us visualize the Decision Tree to understand it better.

```python
#Plot Tree with plot_tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(cls, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

>(The plot_tree returns annotations for the plot, to not show them in the notebook I assigned returned value to _.)

>You can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.

> **For example :**

```python
X_pred = [5.1, 3.2, 1.5, 0.5]
y_pred = cls.predict([X_pred])

print("Prediction is: {}".format(targets[y_pred]))
```
